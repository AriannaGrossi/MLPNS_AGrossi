```python
from __future__ import print_function, division
import pylab as pl
import seaborn
import numpy as np
import scipy as sp
from scipy import stats
import scipy.optimize
%pylab inline
seaborn.set_style("darkgrid")
```

```python
np.random.seed(456)
mymean = 100
```

```python
#Distribuzione del chi quadro
df = mymean
#Distribuzione binomiale
p = 0.5
n= mymean / p #mymean=n*p


mydist = {}
pl.hist(np.random.chisquare(mymean, size=500))
#; alla fine di un comando sopprime l'output della funzione utilizzata
pl.xlabel('x')
pl.ylabel('N');
```

```python
mysize =(np.random.rand(100)*2000).astype(int) 
mysize[mysize<10]=10 #mi assicuro che tutti siano >10
pl.plot(mysize, '.');
#random.rand produce valori floating point tra 0 e 1 in una distribuzione
```

```python
for s in mysize:
  mydist[s]= np.random.chisquare(mymean, size=s)

ax = pl.figure().add_subplot(111)

for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')
ax.plot(pl.xlim(), [100, 100], 'k', alpha=0.5)
ax.set_xlabel("sample size")
ax.set_ylabel("sample mean")
```
