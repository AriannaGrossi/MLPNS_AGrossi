#ESERCIZIO: Il gioco Wordle
Creare un codice che, date due stringhe "guess" e "correct", verifichi se la stringa "guess" è uguale alla stringa "correct" (ossia verifichi che le due stringhe presentino le stesse lettere nelle stesse posizioni), e quindi:

-Se le due stringhe sono uguali, stampi la stringa "DONE"

-Se le due stringhe non sono uguali, stampi la stringa "TRY AGAIN" e stampi una LISTA con un numero di elementi pari alla lunghezza delle due stringhe, e tale per cui: se una lettera della stringa "guess" è presente nella stringa "correct", ma nella posizione sbagliata, l'elemento corrispondente a tale lettera nella lista sia un asterisco, se una lettera della stringa "guess" non è presente nella stringa "correct", l'elemento corrispondente a tale lettera nella lista sia un trattino basso, e se una lettera della stringa "guess" è presente nella stringa "correct" nella posizione giusta, l'elemento corrispondente a tale lettera nella lista sia la lettera stessa

```python
correct ="acres"
output = ""
#"" è una stringa vuota
guess = "test"
for i in range(len(guess)): 
#La funzione len() riceve come argomento una stringa e restituisce la lunghezza 
#del suo argomento
  if correct[i] == guess[i]:
      output= output + correct[i]
  elif guess[i] in correct:
    output= output +"*"
  else:
    output=output + "_"
  
output
```

#Le funzioni
Una singola funzione è costituita da una serie di operazioni, che, quando la funzione viene chiamata, vengono effettuate sull' input inserito dall'utente (oppure a volte vengono effettuate senza bisogno di inserire nessun input). In seguito a tali operazioni la funzione restituisce un "return value".

Sia esistono già delle "built in functions" del linguaggio python (len, range...), sia è possibile creare nuove funzioni.

Per creare una nuova funzione devo fare seguire def dal nome della funzione che voglio creare, seguita a sua volta da una coppia di parentesi contenenti l'input richiesto all'utente (oppure vuote se non è previsto che l'utente inserisca nulla) e dai due punti.

Quindi, tra una coppia di tre virgolette e su righe diverse viene scritta una "docstring", contenente  una descrizione della funzione, l'input e l'output della funzione.

In seguito vengono scritte le operazioni che dovrà compiere la funzione una volta chiamata

```python
def wordle(guess, correct):
  """
  A funcion that checks a word guess against a target
  Input:
    guess: a string
  Result: 
    a string
  """
  test=["_"]*5
  #La lista test è costituita da 5 elementi tutti uguali, ossia da 5 caratteri _
  #(_ viene percepito come carattere perchè è racchiuso tra virgolette)
  #Voglio assicurarmi che tutte le lettere delle due stringhe siano formattate 
  #allo stesso modo
  correct=correct.lower() 
  #Il metodo .lower() non cambia "in place" i caratteri della stringa a cui è 
  #applicato rendendoli tutti minuscoli.
  #Se voglio modificare i caratteri della stringa rendendoli minuscoli dovrò 
  #assegnare alla stringa stessa il "return value" del metodo .lower()
  #(ossia ciò che tale metodo restituisce)
  guess=guess.lower()
  if len(correct)!= len(guess):
    print("wrong word lenght")
    return
  else:
    if correct ==guess:
      print("DONE")
      return
    else:
      print("try again")
      for i in range(len(test)):
        if guess[i]==correct[i]:
          test[i]=guess[i]
        elif guess[i] in correct:
          test[i]="*"
    print("".join(test))
    #La proprietà .join() è associata alle stringhe, e unisce alla stringa a cui
    #è applicata (in questo caso una stringa vuota) il suo argomento (in questo
    #caso gli elementi di una lista)


```

```python
wordle("Acres","Acres")
```

```python
wordle("Acres","Arpal")
```

#ESERCIZIO: PAROLA PALINDROMA

Scrivere una funzione che legga una parola e verifichi se è palindroma: se la parola è palindroma la funzione deve stampare YES, mentre se la parola non è palindroma la funzione deve stampare NO

```python
def ispalyndrom(word):
  """
  Una funzione che verifica se una parola è palindroma
  Input:
   word: una stringa
  Result:
   una stringa
  """
  
  word = word.lower()
  lw = len(word)
  reverseWord = word[::-1]
  #Tratto la stringa word come una array:
  #Poichè scelgo m=-1 inverto gli elementi dell'array word, quindi reverseWord 
  #sarà un array il cui primo elemento sarà l'ultimo elemento dell'array word
  #(ossia l'ultimo carattere della stringa word), mentre l'ultimo elemento sarà
  #lo 0-esimo elemento dell'array word (ossia il primo carattere)
  if word[:int(lw / 2):] == reverseWord[:int(lw/2):]:
  #Poichè il risultato della divisione tra due valori interi è un valore di tipo float,
  #converto tale valore nel valore intero immediatamente inferiore ad esso
  #mediante la funzione int()
  #L'if valuta se la prima metà della stringa word è uguale alla prima metà 
  #della stringa reverseWord 
    print("YES")
  else:
    print("NO")
```
