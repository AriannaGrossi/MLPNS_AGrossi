```python
import numpy as np
import pylab as pl
import scipy
from scipy.spatial.distance import cdist
```

```python
np.random.seed(166)
data= np.random.randn(100,2)
label=np.random.randint(0,2,100)
label
```

```python
data.shape
```

```python
ax=pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')

ax.plot(data[:,0][label==0], data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1], data[:,1][label==1], 's')
ax.set_xlabel("x")
ax.set_xlabel("y")
```

```python
testparticle=np.random.randn(2)
testparticle
```

```python
ax=pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')

ax.plot(data[:,0][label==0], data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1], data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1],'kx',ms=30,lw=3)
ax.set_xlabel("x")
ax.set_xlabel("y")
```

```python
d=cdist(np.atleast_2d(testparticle), data)
d
```

```python
np.min(d)
```

```python
np.argmin(d)
```

```python
np.min(d), d[:,np.argmin(d)]
```

```python
np.argsort(d)
```

```python
d[:,np.argsort(d)]
```

```python
k=1
labeltest=label[np.argmin(d)]
```

```python
ax=pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')
colors=['blue','orange']
ax.plot(data[:,0][label==0], data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1], data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1],'kx',ms=30,lw=3, color=colors[labeltest])
ax.set_xlabel("x")
ax.set_xlabel("y")
```

```python
data[np.argsort(d)][0]
```

```python
k=5
nns=data[np.argsort(d)][0][:k]
nns
```

```python
nnslabel=label[np.argsort(d)][0][:k]
nnslabel
```

```python
nnslabel==0
```

```python
(nnslabel==0).sum()
```

```python
(nnslabel==1).sum()
```

```python
labeltest2 = 0 if (nnslabel == 0).sum() > (nnslabel == 1).sum() else 1
```

```python
ax=pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')
colors=['blue','orange']
ax.plot(data[:,0][label==0], data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1], data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1],'kx',ms=30,lw=3, color=colors[labeltest2])
ax.plot(nns[:,0], nns[:,1], 'kx')
ax.set_xlabel("x")
ax.set_xlabel("y")
```
