#Operazioni tra variabili

Il tipo del risultato di una operazione tra variabili (o di una espressione contenente variabili) verrà ereditato dal tipo della variabile più complessa, in quanto la variabile più complessa è quella che occupa più spazio in memoria




Prodotto di variabili

```python
intVar*floatVar
#posso moltiplicare due variabili dello stesso tipo o di tipo diverso usando *
```

Divisione tra interi

```python
intVar/2
#posso dividere tra loro due variabili usando /,
#dividendo due interi, il risultato della divisione potrà essere sia di tipo int sia di tipo float in base al valore del risultato stesso,
#ossia non avverranno troncamenti ad intero
```

Somma e differenza di variabili

```python
intVar + floatVar
#posso sommare o sottrarre tra loro due variabili dello stesso tipo o di tipo diverso usando + o -
```

Espressioni contenenti variabili

```python
intVar+floatVar-intVar/2
```

```python
type(intVar+floatVar-intVar/2)
#il tipo del risultato di questa espressione viene ereditato dal tipo della variabile più complessa, ossia la variabile floatVar di tipo float
```
